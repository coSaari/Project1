```python
#DataFrame 1
    #Outlays of National Defense by Super Function
        #Primary Columns:
            #Years
            #National Defense Outlay
                #Total outlays in millions of dollars
                    ##Military Personnel, Operation and Maintenance, Procurement, Research, Development, 
                        #Test, and Evaluation
            #Human Resources Outlay
                #Total outlays in millions of dollars
                    ##Education,Training, and Social Services, Health, Medicare, Income Security
                    #Social Security, Vetrans Benefits amd Enviroment, Medicaid and Medicare and General Health allocations
            #Total Federal Outlays
                #Total outlays in millions of dollars
                    #Total Outlays for the federal government per year

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
outlaysDFSupD = outlaySup[["Years","National Defense","Human resources"]].sort_values("Years",ascending=False)
#In pandas set the option(pd.set_option) for the display of maximum columns("display.max_columns"), 
    #allowing dataframe to show every row
#In pandas set the option(pd.set_option) for the display of maximum rows("display.max_rows"), 
    #allowing dataframe to show every row
#create a dataframe for specific columns (years, national defense, human resources)
    #Sort this dataframe on years, from most recent to latest(.sort_values("Years",ascending=False))

outlaysDFSupD["Years"] = outlaySup["Years"].apply(lambda x:x.replace(" estimate",""))
outlaysDFSupD = outlaysDFSupD.drop(outlaysDFSupD.index[0])
outlaysDFSupD["Years"] = outlaysDFSupD["Years"].astype(int)
#apply (.apply) on each item (lambda x:x)in years column in Table 1(outlaySup["Years"]), 
    #replace the string " estimate" with an empty string(.replace(" estimate","")))
#Drop the first row (outlaysDFSupD.index[0])in the DataFrame (outlaysDFSupD)for formatting
#Cast the years column in the dataframe(outlaysDFSupD["Years"]) to an integer(.astype(int)) to plot

outlaysDFSupD["National Defense Outlay"] = outlaySup["National Defense"].apply(lambda x:x.replace(",",""))
outlaysDFSupD["National Defense Outlay"] = outlaysDFSupD["National Defense Outlay"].astype(int)
del outlaysDFSupD["National Defense"]
#apply on each item in the national defense column in Table 1
    #replace the comma with and empty string
#Cast the column at national defense in the new dataframe as an integer to plot
#Delete the previous column at national defense created when making the data frame
    #because we replace with new national defense column with no commas and is cast as an integer
    
outlaysDFSupD["Human Resources Outlay"] = outlaySup["Human resources"].apply(lambda x:x.replace(",",""))
outlaysDFSupD["Human Resources Outlay"] = outlaysDFSupD["Human Resources Outlay"].astype(int)
del outlaysDFSupD["Human resources"]
#apply on each item in the human resources column in Table 1
    #replace the comma with and empty string
#Cast the column at human resources of the new dataframe as an integer to plot
#Delete the previous column in human resources created when making the data frame
    #because we replace with new human resource column with no commas and is cast as an integer

outlaysDFSupD["Total Federal Outlay"] = outlaySup["Total, Federal outlays"].apply(lambda x:x.replace(",",""))
outlaysDFSupD["Total Federal Outlay"] = outlaysDFSupD["Total Federal Outlay"].astype(int)
#apply on each item in the total federal outlay column in Table 1
    #replace the comma with and empty string
#Cast the column at total federal outlay in the new dataframe as an integer to plot
#Didnt need to delete previous column because the total outlays from Table 1 was created after 
    #the creation of the initial dataframe, where it was not called for initialization of the dataframe

outlaysDFSupD
#display the new dataframe for Outlays by Superfunction, with useable data
```

```python
#Figure 1 (DF1)
    #This figure represents the total outlays of the National Defense per year
        #Primary Points of Interest For Legislative Perspective(DoD)
            # For years between 2019 and 2024
                #these points are estimates for future years national defense spending
            #For years until Current
                #Trump Administration in 2016
                    #ISIS resurgence
                #Trump Administration in 2018
                    #Spends $646.2 billion Base DoD and OCO
            # For years between 2003 and 2015 (accounts for m)
                #National defense was impacted by wars in the middle east
                    #2003 Iraq War Starts
                #Bush Administration in 2008
                    #Oversees Contingency Operations Spending Increase
                        #Peaks in 2008 at $186.9 Billion
                        #Sent 20,000 troops to Iraq
                #Obama Administration in 2009
                    #Sends 47,000 troops to Afghanistan
                #Obama Administration 2010
                    #Spends $691 Billion for Base DoD and OCO
                #Obama Administration in 2011
                    #Ends war in Iraq
                #Obama Administration in 2012 
                    #hit high DoD base busget $530 billion 
                    #Troop withdrawl from Afghanistan, costs fall
                #Obama Administration in 2013
                    #Cut significant troop involvement in Afghanistan
                        #by 2016 only 12,457 boots on the ground in middle east

        #Key Notes for understanding Tables
            #Overall Structure of function creation
                #The functional structure divides the budget into 18 broad
                    #areas (functions) that provide a coherent and comprehensive basis for analysis.
                #Each function, in turn, is divided into basic groupings of programs, called subfunctions.
            #Arraying Data on a functional basis
                #Budget authority and outlays are classified according to the primary purpose of the activity.
                    #this classification is made without regard to agency or organizational distinctions.
                #Classifying each activity solely in the function defining its most important purpose.
                        #Ex: 
                            # Federal spending for Medicaid constitutes a health care program, but it also 
                        #constitutes a form of income security benefits. However,the spending cannot be counted in both 
                        #functions; since the main purpose of Medicaid is to finance the health care of the beneficiaries, 
                        #this program is classified in the healthfunction.
                #The outlays by agency are based on the agency structure currently in effect.
                        #Ex:
                            # the Department of Homeland Security was established by legislation enacted in
                            #2002. However, these data show spending by the Department of Homeland Security in
                            #previous years that consists of spending attributable to predecessor agencies in earlier years,
                            #but now attributable to the Department of Homeland Security.
            #SuperFunction breakdown
            #Years
            #National Defense Outlay
                #Percentage of total outlays out of total federal outlays(1)
                    #Department of Defense-Military
                        #Military Personnel, Operation and Maintenance, Procurement, Research, Development, 
                        #Test, and Evaluation
            #Human Resources Outlay
                #Percentage of total outlays out of total federal outlays(1)
                    #Education,Training, and Social Services, Health, Medicare, Income Security
                    #Social Security, Vetrans Benefits amd Enviroment, Medicaid and Medicare and General Health allocations
            #Physical Resources Outlay
                #Percentage of total outlays out of total federal outlays(1)
                    #Energy, Natural Resources and Enviroment, Commerce and House Credit**, Transportation,
                    #Community and Regional Development
            #Total Federal Outlays
                #Percentage of total outlays out of total federal outlays(1)
                    #Total Outlays for the federal government per year
                
                
import matplotlib.pyplot as plt
natDPlot = sb.factorplot(x = "Years",y = "National Defense Outlay",data=outlaysDFSupD,size=10)
natDPlot.set_xticklabels(rotation=90)


plt.title("Primary Outlays Per Year", fontsize = 15)
plt.xlabel("Year", fontsize = 10) # label for x-axis
plt.ylabel("Total Outlays (millions)", fontsize = 15)


#Create a new seaborn factor plot that plots according to years on the x axis
    #and national defense outlay, uses data from DataFrame 1(table 1), and increses the size of the plot to allow 
    #x ticks to be visible
#along with spacing solved by size above, set_xticklabels allows for rotation to be passed as an integer parameter
    # and when set to 90 will rotate to vertical
    
#Using matplotlib to add a title and x / y labels
```
